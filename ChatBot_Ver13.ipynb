```python
import tkinter as tk
import tkinter.font as font
import os
from time import sleep

class Application(tk.Frame):
    """メインウィンドウ作成クラス

        フレームを作成し、画面の各種設定を行う。
        フレーム上にラベルやボタンなどのウィジェットを配置し、メインウィンドウを作成する。
        画面に表示する部品に対する処理や部品を追加する際などは、このクラスを修正する。

        Attributes:
            master (Tkinter): メインウィンドウ
            

    """
    def __init__(self, master):
        # スーパークラスFrameの__init__() を呼び出して初期化に必要な処理を実施
        super().__init__(master)
        # フレームをメインウィンドウに配置し
        self.pack()
        # 画面サイズの設定
        self.master.geometry("470x630")
        # TODO タイトルの設定
        self.master.title("犬チャットボットve05_4")
        # フォントの設定
        self.master.option_add("*font", ["MSゴシック", 24])
        # 部品の配置
        self.create_widgets()
        # ChatControlのインスタンス
        self.chatControl = ChatControl()

    def create_widgets(self):
        """ウィジェットの設定

            メインウィンドウにキャンバスを設定し、ウィジェットを設置する。
            表示する画像を変更する場合は、このメソッドを修正する。
            
            Attributes:
                canvas:(:obj:Canvas):キャンバス（ウィジェットを設置する土台）のオブジェクト
                img_list (list): 画面に表示する画像のリスト
                frame_img:(:obj:PhotoImage):画像の設定を行うオブジェクト
                askbutton:(:obj:Button):ボタンのオブジェクト
                answer_font:(:obj:Font):フォント
                answer:(:obj:Label):ラベル
                inputarea:(:obj:Entry):入力フィールド
        
        """
        # 画像表示
        self.canvas = tk.Canvas(bg="white", width=470, height=630)
        self.canvas.place(x=0, y=0)
        
        # TODO チャットボットイメージ画像の設定（表示画像拡張エリア）
        self.img_list = []
        self.img_list.append(tk.PhotoImage(file="images/dog00.png"))
        self.img_list.append(tk.PhotoImage(file="images/dog02.png"))
        self.img_list.append(tk.PhotoImage(file="images/dog03.png"))
        self.img_list.append(tk.PhotoImage(file="images/dog04.png"))
        self.img_list.append(tk.PhotoImage(file="images/dog05.png"))
        self.canvas.create_image(235,280,image=self.img_list[0],tag="ai-bot")
        # フレームに画像の設定
        self.frame_img = tk.PhotoImage(file="images/tablet.png")
        self.canvas.create_image(235, 315, image=self.frame_img)
        
        #ボタン
        self.askbutton = tk.Button(text="▲", font=("", 18))
        self.askbutton.place(x=386, y=511)
        #ボタンに紐づけ
        self.askbutton["command"] = self.ask_click

        #出力エリア
        self.canvas.create_rectangle(40, 420, 430, 500, outline = "pink", fill = "white")
        self.answer_font = font.Font(self.master,family="HG創英角ﾎﾟｯﾌﾟ体",size=16)
        self.answer = tk.Label(text="(話しかけられるのを待ってる)", bg="white", font=self.answer_font)
        self.answer.place(x=50, y=430)

        #入力エリア
        self.inputarea = tk.Entry(width=21, bd=4)
        self.inputarea.place(x=34, y=511)
        #リターンキー対応
        self.inputarea.bind('<Return>', self.ask_enter)        

    #エンターキー押した
    def ask_enter(self,event):
        self.ask_click()

    #ボタン押した
    def ask_click(self):
        """ボタン押下時の動作制御

            入力メッセージを取得し実行する処理の制御を行う。
            ChatControlクラスのメソッドからの戻り値にしたがって、以下の処理を行う。
            ・応答メッセージの設定
            ・画面表示画像の切り替え
            ・画面の初期化
        
        """
        # 入力値の取得
        message = self.inputarea.get()
        # 入力欄クリア
        self.inputarea.delete(0, tk.END)
        
        # チャット制御
        chat = self.chatControl.control(message)
        replay_message = chat.replay_message
        image_no = chat.image_idx
        init_flg = chat.init_flg
        shutdown_flg=chat.shutdown_flg
        
        self.answer["text"] = "connecting"
        for i in range(3):
            self.master.update()
            self.answer["text"] += "."
            sleep(0.5)
        
        self.canvas.itemconfig("ai-bot", image=self.img_list[image_no])
        self.answer["text"] = replay_message
        self.master.update()
        if init_flg:
            sleep(3)
            self.canvas.itemconfig("ai-bot", image=self.img_list[0])
            self.answer["text"] = "話しかけられるのを待ってる。"
        if shutdown_flg:
            os.system('shutdown -s -r')
        
```

```python
class Chat():
    """チャットクラス

        チャットの応答情報を管理する
        
        Attributes:
            replay_message(str): 応答メッセージ
            image_idx(int):出力画像のidex
            init_flg(bool):初期化flg
            shutdown_flg:シャットダウンflg
    """
    
    # TODO
    def set_replay_data(self,replay_message,image_idx=0,init_flg=False,shutdown_flg=False):
        self.replay_message = replay_message
        self.image_idx = image_idx
        self.init_flg = init_flg
        self.shutdown_flg=shutdown_flg
```

```python
class ChatControl():
    """チャットコントロールクラス

        チャットの処理を制御する
        
        Attributes:
            chat(:obj:Chat): チャットの応答情報
            chatLogic(:obj:ChatLogic):チャットの各機能の処理
    """
    def __init__(self):
        self.chatLogic = ChatLogic()
    
    def control(self,message):
        """チャットの制御を行う
        
            機能の実行状態に従って処理を振り分ける。
            機能実行状態のフラグがTrueであれば、該当機能を実行する。
            機能実行状態のフラグが全てFalseであれば、replayメソッドを実行する。
        
            Args:
                message (str): 処理対象のメッセージ.
                
            Returns:
                obj:Chat: チャットの応答情報
            Raise:
                ValueError
        
        """
        # 機能継続チェック
        # TODO
        chat=Chat()
        info =Info()
        print(self.chatLogic.status.del_event_flg)
        if self.chatLogic.status.calc_flg:
            try:
                chat=self.chatLogic.calc_func(message)
            except ValueError as e:
                self.chatLogic.status.calc_flg=False
                self.chatLogic.status.payroll_flg=False
                chat.set_replay_data('数字以外はダメ やめるね！',4,True)
        elif self.chatLogic.status.hitgame_flg:
            
            chat=self.chatLogic.hitgame_func(message)
                
        elif self.chatLogic.status.info_flg:
                chat=info.choise_word(message)
        elif self.chatLogic.status.event_flg:
            chat=self.chatLogic.event_func(message)
        elif self.chatLogic.status.del_event_flg:
            print('del_event')
            chat=self.chatLogic.del_event_func(message)
        elif self.chatLogic.status.salary_flg:
            chat=self.chatLogic.predict_salary_func(message)
        elif self.chatLogic.status.ct_flg:
            chat=self.chatLogic.predict_ct_func(message)
        else:
            chat=self.chatLogic.replay(message)
                
        return chat
```

```python
class PredictSalary():
    """
    年収予測クラス
    年収予測に必要な情報と処理を定義する
    年収予測の学習、学習済みモデルの保存、読み込み、年収予測を行う。
    attributes:
        model(obj:LinearRegression):学習モデル
        model_file(str):学習済みモデルのファイルパス
        train_file(str):学習用データのファイルパス
        predict_flg(bool):予測実行フラグ
    """
    def __init__(self):
        
        self.model=None
        self.model_file='data/salary_model.pkl'
        self.train_file='data/salary.csv'
        self.predict_flg=False
    def save_model(self):
        """学習済みモデル保存
        学習済みモデルを保存する
        """
        import pickle
        #    学習済みモデルの保存
        with open(self.model_file,'wb')as file:
            pickle.dump(self.model,file)
    def read_model(self):
        """
        学習済みモデル読み込み
        学習済みモデルを読み込む
        """
        import pickle
        with open(self.model_file,"rb") as file:
            self.model=pickle.load(file)
            return self.model
    def learn(self):
        """
        学習
        機械学習の回帰を行うアルゴリズムを用いて給料予測データを学習する
        学習が終わったら、学習済みモデルを保存する
        """
        import pandas as pd
        from sklearn import linear_model 
        # データ解析用ライブラリの読み込み
        
        # 指定列の抽出
        train=pd.read_csv(self.train_file)
        # 機械学習モデルを準備する
        self.model =linear_model.LinearRegression()

        # 学習データから問題（説明変数）を抽出する
        data_in=train.drop('salary',axis=1)

        # 学習データから答え（目的変数）を抽出する
        data_out=train['salary']
        # 機械学習を実行する
        self.model.fit(data_in,data_out)
        self.save_model()
    def predict(self,age):
        """予測
        引数で指定された年齢をもとに年収の予測結果を返す。
        予測は、学習済みモデルを読み込み行う
        Args:
            age(int): 予測を行う年齢
        Returns:
            str:ファイルから読み込んだ内容
            float:予測結果の年収
        """
        import pandas as pd
        from sklearn import linear_model 
        import matplotlib.pyplot as plt
        trained_model=self.read_model()
        test=pd.DataFrame({"age":[age]})
        # ラベルの定義
        # テスト実行
        result=trained_model.predict(test)
        return result[0]
```

```python
class Calc():
    """計算クラス

        計算対象の値を保持し、呼び出されたメソッドに応じて計算をする。
                
        Attributes:
            x(int): 計算対象の値
            y(int): 計算対象の値
    """
    
    def __init__(self):
        self.x = None
        self.y = None

    def calc_add(self):
        """2つの値の足し算を行う
        
        Returns:
            int: 加算の結果
        """
        
        return self.x + self.y
    def calc_mul(self):
        return self.x* self.y
```

```python
class Event:
    
    """
    イベント情報を管理する
    イベントまでの日数取得やイベントの登録を行う
    attributes:
        event_name(str):イベント名
        event_date(str): イベントの日程
    """
    
    def __init__(self):
        self.event_name=None
        self.event_date=None
        self.file_path='data/event.txt'
    def get_util_date(self,message):
        """
        イベントまでの日数取得
        
        指定されたイベントまでの残り日数を計算して返す。
        入力されているメッセージは、「○○まであと何日」の想定なので、
        ○○のイベント名を抽出する。
        日付情報は、「YYYY/MM/DD」形式でファイルに格納されているので、
        適切に取得する。
        現在日時とイベントの日程の差分を求めて結果を返す。
        Returns:
            str:イベントまでの残り日数
        """
        import datetime as dt
        file=FileUtil()
        event_name=message.replace('まであと何日','')
        event=file.json_file_read(self.file_path)
        print('event30')
        event_date=event[event_name]
        event_date=dt.datetime.strptime(event_date,'%Y/%m/%d')
        now_date=datetime.datetime.now()
        date=event_date-now_date
        return str(date.days+1)
    def add_event(self):
        file=FileUtil()
        event1=file.json_file_read(self.file_path)
        event2={self.event_name:self.event_date}
        event1.update(event2)
        file.json_file_write('data/event.txt',event1)
    def del_event(self,del_event_name):
        file=FileUtil()
        event=file.json_file_read(self.file_path)
        del event[del_event_name]
    def date_format_chack(self,message):
        import re
        result=re.fullmatch(r'\d{4}/\d{2}/\d{2}', message)
        if result!=None:
            pass
        else:
            raise FormatError
```

```python
class FormatError(Exception):
    pretxt=''
    def __init__(self, message,*args):
        if self.pretxt:
            message = f"{self.pretxt}:{message}"
        super().__init__(message,*args)
class TextFormatError(FormatError):
    pretxt='形式が不正'
```

```python

class FileUtil:
    """
    様々なファイル操作の機能を提供するクラス。
    """
    def file_read(self,file_path):
        """
        ファイル読み込み
        引数で指定されたファイルをオープンして内容を読み込む。
        args:
            file_path(str):読み込み対象のファイル
        Returns:
            str:ファイルから読み込んだ内容
        Raises:
            IoError:ファイルが存在しない場合に発生
        """
        with open(file_path, 'r',encoding='utf-8') as file:
            line=file.read()
            
            return line
    def file_write(self,file_path,value):
        """
        ファイル書き込み
        引数で指定されたファイルに引数で受け取った値を書き込む
        args:
            file_path(str):書き込み対象のファイル
            value(str):書き込む値
        Returns:
            str:ファイルから読み込んだ内容
        Raises:
            IoError:ファイルをほかのプロセスで使用している場合に発生
        """
        with open(file_path,'w',encoding='utf-8') as file:
            file.write(value)
            file.flush()
    def json_file_read(self,file_path):
        """
        JSON形式のファイル読み込み
        引数で指定されたJSON形式のファイルをオープンし内容を読み込む
        Args:
        file_path(str):読み込み対象のファイル
        Returns:
            str:ファイルから読み込んだ内容
        Raises:
            IoError:ファイルが存在しない場合に発生
            JSONEncodeError:データが０件の時にloadすると発生
        """
        import json
        with open(file_path, 'r',encoding='utf-8') as file:
            self.line={}
            try:
                self.line=json.load(file)
            except json.decoder.JSONDecodeError as e:
                print(e)
            return self.line
    def json_file_write(self,file_path,dic_date):
        import json
        """
        JSON形式のファイル読み込み
        引数で指定された辞書型のデータをjson形式でファイルに書き込む
        args:
            file_path(str):書き込み対象のファイル
            dic_date(str):書き込む値
        Raises:
            IoError:ファイルをほかのプロセスで使用している場合に発生
        """
        with open(file_path,'w',encoding='utf-8') as file:
            json.dump(dic_date, file,ensure_ascii=False)
            file.flush()
```
